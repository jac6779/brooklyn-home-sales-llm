# Brooklyn Home Sales – Export Pipeline

This notebook packages preprocessing and modeling components into a single reusable pipeline for predicting Brooklyn home sale prices.

### Goals
- Load trained preprocessing and model artifacts
- Combine them into a unified pipeline
- Export the pipeline for downstream use (e.g., APIs, batch scoring)

### Outputs
- Serialized `.joblib` pipeline ready for production use

### Imports & Dependencies
This section loads all required libraries for:
- Data manipulation
- Model handling
- Pipeline construction
- File system operations

```python
# Import libraries
import joblib
import numpy as np
import pandas as pd
import sys
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
```

```python
# File path locations
from pathlib import Path

# File paths
PROJECT_ROOT = Path.cwd().resolve().parent
CLEAN_DATA_DIR = PROJECT_ROOT / "data/clean_data"
RAW_DATA_DIR = PROJECT_ROOT / "data/raw_data"
MODELS_DIR = PROJECT_ROOT / "models"
SRC_DIR = PROJECT_ROOT / "src"
DEPLOYMENT_DIR = PROJECT_ROOT / "deployment"

sys.path.append(str(PROJECT_ROOT))

from app.feature_builder import build_model_features
```

### Load Preprocessing & Model Artifacts

In this section, I load previously trained components:

- **Preprocessor**: Handles feature transformations (scaling, encoding, etc.)
- **Model**: Trained regression model for predicting home prices

These artifacts are loaded using `joblib`.

```python
# Load artifacts
preprocessor = joblib.load(MODELS_DIR / "preprocessing_pipeline.joblib")
model = joblib.load(MODELS_DIR / "home_price_model.joblib")

# Combine artifacts into single pipeline
raw_input_pipeline = Pipeline([
    ("feature_builder", FunctionTransformer(build_model_features, validate=False)),
    ("preprocessor", preprocessor),
    ("model", model)
])
```

### Export Final Pipeline

The final pipeline is serialized and saved to disk.

#### Output:
- `brooklyn_price_pipeline_raw_inputs.joblib`

#### Use cases:
- Batch scoring jobs
- Real-time prediction APIs
- Integration into other systems

This file can be loaded and used without retraining.

```python
# Save combined pipeline
joblib.dump(raw_input_pipeline, MODELS_DIR / "brooklyn_price_pipeline_raw_inputs.joblib")
```
